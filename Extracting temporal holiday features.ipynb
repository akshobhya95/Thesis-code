```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame with columns 'collection_date' and 'bed_occupancy_rate'

# Convert 'collection_date' to datetime if it's not already in datetime format
df['collection_date'] = pd.to_datetime(df['collection_date'])

# Extract temporal features
df['day_of_week'] = df['collection_date'].dt.dayofweek  # Day of the week (Monday = 0, Sunday = 6)
df['day_of_month'] = df['collection_date'].dt.day       # Day of the month (1-31)
df['month'] = df['collection_date'].dt.month            # Month (1-12)
df['quarter'] = df['collection_date'].dt.quarter        # Quarter (1-4)
df['year'] = df['collection_date'].dt.year             # Year
# You can extract more features like season, is_weekend, is_holiday, etc., if needed

# Cyclic feature engineering
df['day_of_year_sin'] = np.sin(2 * np.pi * df['collection_date'].dt.dayofyear / 365)
df['day_of_year_cos'] = np.cos(2 * np.pi * df['collection_date'].dt.dayofyear / 365)
df['month_sin'] = np.sin(2 * np.pi * df['collection_date'].dt.month / 12)
df['month_cos'] = np.cos(2 * np.pi * df['collection_date'].dt.month / 12)
df['day_of_month_sin'] = np.sin(2 * np.pi * df['collection_date'].dt.day / 31)
df['day_of_month_cos'] = np.cos(2 * np.pi * df['collection_date'].dt.day / 31)
df['day_of_week_sin'] = np.sin(2 * np.pi * df['collection_date'].dt.dayofweek / 7)
df['day_of_week_cos'] = np.cos(2 * np.pi * df['collection_date'].dt.dayofweek / 7)

# Extract season as numerical values
def get_season(month):
    if month in [3, 4, 5]:
        return 1  # Spring
    elif month in [6, 7, 8]:
        return 2  # Summer
    elif month in [9, 10, 11]:
        return 3  # Autumn
    else:
        return 4  # Winter

df['season'] = df['month'].apply(get_season)

# Extract is_weekend
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# Now you have added temporal features, cyclic features, season (as numerical values), and is_weekend to your DataFrame 'df'
```

```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame with the features to be standardized

# Function to perform quantile standardization on a single column
def quantile_standardization(column):
    # Rank the data
    ranked_values = column.rank(method='average', pct=True)
    # Compute the quantile function
    quantiles = pd.Series(np.quantile(np.random.normal(size=len(column)), ranked_values), index=column.index)
    # Replace original values with quantiles
    standardized_values = quantiles.sort_values()
    return standardized_values

# Apply quantile standardization to each column, excluding the first three columns
df_quantile_standardized = df.iloc[:, 3:].apply(quantile_standardization, axis=0)

# Concatenate the first three columns with the quantile standardized values
df_quantile_standardized = pd.concat([df.iloc[:, :3], df_quantile_standardized], axis=1)

# Display the quantile standardized DataFrame
print(df_quantile_standardized)
```
