```python
import pandas as pd



# Read the CSV file into a pandas DataFrame
df = pd.read_csv('covid-19_stateextraction.csv',delimiter=';')

# Display the DataFrame
df
```

```python
import pandas as pd

# Define the region names and their corresponding states
region_states = {
    "Region 1": "Boston (Connecticut, Maine, Massachusetts, New Hampshire, Rhode Island, Vermont)",
    "Region 2": "New York (New Jersey, New York, Puerto Rico, Virgin Islands)",
    "Region 3": "Philadelphia (Delaware, District of Columbia, Maryland, Pennsylvania, Virginia, West Virginia)",
    "Region 4": "Atlanta (Alabama, Florida, Georgia, Kentucky, Mississippi, North Carolina, South Carolina, Tennessee)",
    "Region 5": "Chicago (Illinois, Indiana, Michigan, Minnesota, Ohio, Wisconsin)",
    "Region 6": "Dallas (Arkansas, Louisiana, New Mexico, Oklahoma, Texas)",
    "Region 7": "Kansas City (Iowa, Kansas, Missouri, Nebraska)",
    "Region 8": "Denver (Colorado, Montana, North Dakota, South Dakota, Utah, Wyoming)",
    "Region 9": "San Francisco (Arizona, California, Hawaii, Nevada, American Samoa, Northern Mariana Islands, Federated States of Micronesia, Guam, Marshall Islands, Palau)",
    "Region 10": "Seattle (Alaska, Idaho, Oregon, Washington)"
}

# Fill in the 'state' column based on the 'jurisdiction' column
for region, states in region_states.items():
    df.loc[df['jurisdiction'] == region, 'state'] = states

print(df)
```

```python
df.to_csv('covid-19_stateextraction.csv',index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the columns of interest
columns_of_interest = ['collection_date', 
                       'percent_inpatient_beds_occupied', 
                       'percent_staff_icu_beds_occupied', 
                       'total_admissions_all_influenza_confirmed_past_7days', 
                       'total_admissions_all_covid_confirmed_past_7days', 
                       'total_patients_hospitalized_covid_confirmed', 
                       'total_patients_hospitalized_influenza_confirmed']

# Create a DataFrame with the selected columns
df_filtered = df[columns_of_interest]

# Convert 'collection_date' to datetime
df_filtered['collection_date'] = pd.to_datetime(df_filtered['collection_date'])

# Melt the DataFrame to long format for easier plotting
df_melted = df_filtered.melt(id_vars='collection_date', 
                              var_name='Variable', 
                              value_name='Value')

# Plot using seaborn
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_melted, x='collection_date', y='Value', hue='Variable')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Trends Over Time')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Assuming your DataFrame is named df and contains the relevant columns

# Filter relevant columns
columns_of_interest = ['collection_date', 
                       'percent_inpatient_beds_occupied', 
                       'percent_staff_icu_beds_occupied', 
                       'total_admissions_all_influenza_confirmed_past_7days', 
                       'total_admissions_all_covid_confirmed_past_7days', 
                       'total_patients_hospitalized_covid_confirmed', 
                       'total_patients_hospitalized_influenza_confirmed']
df_filtered = df[columns_of_interest]

# Convert 'collection_date' to datetime if it's not already
df_filtered['collection_date'] = pd.to_datetime(df_filtered['collection_date'])

# Aggregate data by day
df_filtered['date'] = df_filtered['collection_date'].dt.date
df_aggregated = df_filtered.groupby('date').mean().reset_index()

# Plot bed occupancy rates over time
plt.figure(figsize=(12, 8))
plt.plot(df_aggregated['date'], df_aggregated['percent_inpatient_beds_occupied'], label='Inpatient Beds Occupancy')
plt.plot(df_aggregated['date'], df_aggregated['percent_staff_icu_beds_occupied'], label='ICU Beds Occupancy')
plt.xlabel('Date')
plt.ylabel('Occupancy Rate (%)')
plt.title('Daily Average Metrics of Hospitalization Trends of Bed Occupancy Rates')
plt.legend()
plt.grid(True)
plt.show()

# Plot COVID-19 and influenza admissions over time
plt.figure(figsize=(12, 8))
plt.plot(df_aggregated['date'], df_aggregated['total_admissions_all_influenza_confirmed_past_7days'], label='Influenza Admissions')
plt.plot(df_aggregated['date'], df_aggregated['total_admissions_all_covid_confirmed_past_7days'], label='COVID-19 Admissions')
plt.xlabel('Date')
plt.ylabel('Number of Admissions')
plt.title('Daily Average Metrics of Hospitalization Trends  of Influenza and COVID-19 Admissions')
plt.legend()
plt.grid(True)
plt.show()

# Calculate proportion of inpatient beds taken by COVID-19 patients compared to all patients
df_aggregated['covid_proportion'] = df_aggregated['total_patients_hospitalized_covid_confirmed'] / (df_aggregated['total_patients_hospitalized_covid_confirmed'] + df_aggregated['total_patients_hospitalized_influenza_confirmed'])

# Plot proportion of inpatient beds occupied by COVID-19 patients over time
plt.figure(figsize=(12, 8))
plt.plot(df_aggregated['date'], df_aggregated['covid_proportion'])
plt.xlabel('Date')
plt.ylabel('Proportion')
plt.title('Daily Average Metrics of Hospitalization Trends propotion of Inpatient Beds Occupied by COVID-19 Patients')
plt.grid(True)
plt.show()

```
