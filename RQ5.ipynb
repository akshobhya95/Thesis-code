```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from mapie.regression import MapieRegressor
from mapie.metrics import regression_coverage_score
import matplotlib.pyplot as plt

# Select relevant columns for analysis
selected_columns = ['collection_date', 'total_hospitals', 'admissions_all_covid_confirmed', 
                    'total_patients_hospitalized_covid_confirmed', 'inpatient_beds', 
                    'total_icu_beds', 'day_of_week', 'month', 'season', 'is_weekend',
                    'day_of_year_sin', 'day_of_year_cos']

# Assuming df is already loaded with the data and selected_columns are correctly defined
df = df[selected_columns].copy()

# Example: Engineering sine and cosine transformations for seasonal components
df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)

df['day_of_year_sin'] = np.sin(2 * np.pi * df['day_of_year_sin'])
df['day_of_year_cos'] = np.cos(2 * np.pi * df['day_of_year_cos'])

df['day_of_week_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['day_of_week_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)

# Drop original categorical columns after encoding
df.drop(['day_of_week', 'month'], axis=1, inplace=True)

# Split data into features (X) and target variable (y)
X = df.drop(['collection_date', 'admissions_all_covid_confirmed'], axis=1)
y = df['admissions_all_covid_confirmed']

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Initialize Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model
rf_model.fit(X_train, y_train)

# Predict on test set
y_pred = rf_model.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)

# Generate prediction intervals using Conformal Prediction with ACI
mapie = MapieRegressor(estimator=rf_model, method="plus")
mapie.fit(X_train, y_train)
y_pred, y_pis = mapie.predict(X_test, alpha=0.1)

# Ensure y_pis is 2-dimensional
if len(y_pis.shape) == 1:
    y_pis = y_pis.reshape(-1, 2)

# Ensure prediction intervals are sorted
y_pis_sorted = np.sort(y_pis, axis=1)

# Evaluate coverage
coverage = regression_coverage_score(y_test, y_pis[:, 0], y_pis[:, 1])

# Calculate median lower and upper bounds
median_lower_bound = np.median(y_pis[:, 0])
median_upper_bound = np.median(y_pis[:, 1])

# Calculate IQR lower and upper bounds
iqr_lower_bound = np.percentile(y_pis[:, 0], 25)
iqr_upper_bound = np.percentile(y_pis[:, 1], 75)

# Calculate median width of prediction intervals
median_width = np.median(y_pis[:, 1] - y_pis[:, 0])

# Calculate MAPE and Median APE
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
median_ape = np.median(np.abs((y_test - y_pred) / y_test)) * 100

# Output results
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Median Lower Bound: {median_lower_bound:.2f}')
print(f'Median Upper Bound: {median_upper_bound:.2f}')
print(f'IQR Lower Bound: {iqr_lower_bound:.2f}')
print(f'IQR Upper Bound: {iqr_upper_bound:.2f}')
print(f'Median Width: {median_width:.2f}')
print(f'MAPE: {mape:.2f}%')
print(f'Median APE: {median_ape:.2f}%')
print(f'Coverage: {coverage:.2f}')

# Plot prediction intervals
plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label='True values', color='blue')
plt.plot(y_pred, label='Predicted values', color='green')

# Ensure y_pis_sorted is 2-dimensional
if y_pis_sorted.ndim == 2:
    plt.fill_between(np.arange(len(y_test)), y_pis_sorted[:, 0], y_pis_sorted[:, 1], color='gray', alpha=0.5, label='Prediction Interval')


plt.xlabel('Time')
plt.ylabel('COVID-19 Admissions')
plt.title('Prediction Intervals for COVID-19 Admissions')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from mapie.regression import MapieRegressor
import matplotlib.pyplot as plt

# Load and preprocess your data
# df = pd.read_csv('your_data.csv')

# Ensure collection_date is in datetime format
df['collection_date'] = pd.to_datetime(df['collection_date'], format='%d-%m-%Y')

# Define your feature sets
features_with_hospital_metrics = [
    'admissions_70_covid_confirmed', 'admissions_60_69_covid_confirmed', 
    'average_admissions_30_49_covid_confirmed_per_100k', 
    'percent_inpatient_beds_occupied', 
    'percent_change_total_admissions_all_influenza_confirmed_past_7days_per_100k',
    'peak_flu_season', 'is_holiday?', 'quarter', 'day_of_year_sin', 
    'day_of_year_cos', 'month_sin', 'month_cos', 'day_of_month_sin', 
    'day_of_month_cos', 'day_of_week_sin', 'day_of_week_cos', 'season', 
    'is_weekend'
]
features_without_hospital_metrics = [
    'peak_flu_season', 'is_holiday?', 'quarter', 'day_of_year_sin', 
    'day_of_year_cos', 'month_sin', 'month_cos', 'day_of_month_sin', 
    'day_of_month_cos', 'day_of_week_sin', 'day_of_week_cos', 'season', 
    'is_weekend'
]

# Prepare your data
X_with_hospital_metrics = df[features_with_hospital_metrics]
X_without_hospital_metrics = df[features_without_hospital_metrics]
y = df['admissions_all_covid_confirmed']
dates = df['collection_date']

# Split data into training and test sets
X_train_with, X_test_with, y_train, y_test, dates_train, dates_test = train_test_split(X_with_hospital_metrics, y, dates, test_size=0.2, shuffle=False)
X_train_without, X_test_without, y_train, y_test, dates_train, dates_test = train_test_split(X_without_hospital_metrics, y, dates, test_size=0.2, shuffle=False)

# Initialize and train your models
rf_model_with = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_without = RandomForestRegressor(n_estimators=100, random_state=42)

# Train MapieRegressor for prediction intervals without shuffling
mapie_with = MapieRegressor(estimator=rf_model_with, method="plus")
mapie_without = MapieRegressor(estimator=rf_model_without, method="plus")

mapie_with.fit(X_train_with, y_train)
mapie_without.fit(X_train_without, y_train)

# Obtain predictions and prediction intervals
y_pred_with, y_pis_with = mapie_with.predict(X_test_with, alpha=0.1)
y_pred_without, y_pis_without = mapie_without.predict(X_test_without, alpha=0.1)

# Ensure y_pis_with and y_pis_without are the correct shape
if len(y_pis_with.shape) == 3:
    y_pis_with = y_pis_with[:, :, 0]
if len(y_pis_without.shape) == 3:
    y_pis_without = y_pis_without[:, :, 0]

# Debugging: Print shapes and first few values
print(f'y_pis_with reshaped: {y_pis_with.shape}, first values: {y_pis_with[:5]}')
print(f'y_pis_without reshaped: {y_pis_without.shape}, first values: {y_pis_without[:5]}')

# Calculate coverage manually
def calculate_coverage(y_true, y_lower, y_upper):
    coverage = np.mean((y_true >= y_lower) & (y_true <= y_upper))
    print(f'Coverage calculation - y_true: {y_true[:5]}, y_lower: {y_lower[:5]}, y_upper: {y_upper[:5]}, coverage: {coverage}')
    return coverage

def calculate_width(y_lower, y_upper):
    width = np.mean(y_upper - y_lower)
    print(f'Width calculation - y_lower: {y_lower[:5]}, y_upper: {y_upper[:5]}, width: {width}')
    return width

# Calculate coverage and width
coverage_with = calculate_coverage(y_test, y_pis_with[:, 0], y_pis_with[:, 1])
coverage_without = calculate_coverage(y_test, y_pis_without[:, 0], y_pis_without[:, 1])

width_with = calculate_width(y_pis_with[:, 0], y_pis_with[:, 1])
width_without = calculate_width(y_pis_without[:, 0], y_pis_without[:, 1])

# Calculate evaluation metrics
mae_with = mean_absolute_error(y_test, y_pred_with)
mae_without = mean_absolute_error(y_test, y_pred_without)

rmse_with = np.sqrt(mean_squared_error(y_test, y_pred_with))
rmse_without = np.sqrt(mean_squared_error(y_test, y_pred_without))

r2_with = r2_score(y_test, y_pred_with)
r2_without = r2_score(y_test, y_pred_without)

mad_with = np.median(np.abs(y_test - y_pred_with))
mad_without = np.median(np.abs(y_test - y_pred_without))

mape_with = np.mean(np.abs((y_test - y_pred_with) / y_test)) * 100
mape_without = np.mean(np.abs((y_test - y_pred_without) / y_test)) * 100

median_ape_with = np.median(np.abs((y_test - y_pred_with) / y_test)) * 100
median_ape_without = np.median(np.abs((y_test - y_pred_without) / y_test)) * 100

# Output results
print(f'Model with Hospital Metrics - MAE: {mae_with:.2f}, RMSE: {rmse_with:.2f} ,R2: {r2_with:.2f}, MAD: {mad_with:.2f}, MAPE: {mape_with:.2f}%, Median APE: {median_ape_with:.2f}%, Coverage: {coverage_with:.2f}, Width: {width_with:.2f}')
print(f'Model without Hospital Metrics - MAE: {mae_without:.2f}, RMSE: {rmse_without:.2f}, R2: {r2_without:.2f}, MAD: {mad_without:.2f}, MAPE: {mape_without:.2f}%, Median APE: {median_ape_without:.2f}%, Coverage: {coverage_without:.2f}, Width: {width_without:.2f}')

# Plot comparison
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 8), sharey=True)

# Plot with hospital metrics
axes[0].plot(dates_test, y_test.values, label='True values', color='blue')
axes[0].plot(dates_test, y_pred_with, label='Predicted values with Hospital Metrics', color='green')
axes[0].fill_between(dates_test, y_pis_with[:, 0], y_pis_with[:, 1], color='green', alpha=0.2, label='Prediction Interval with Hospital Metrics')
axes[0].set_title('Predictions with Hospital Metrics')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('COVID-19 Admissions')
axes[0].legend()

# Plot without hospital metrics
axes[1].plot(dates_test, y_test.values, label='True values', color='blue')
axes[1].plot(dates_test, y_pred_without, label='Predicted values without Hospital Metrics', color='red')
axes[1].fill_between(dates_test, y_pis_without[:, 0], y_pis_without[:, 1], color='red', alpha=0.2, label='Prediction Interval without Hospital Metrics')
axes[1].set_title('Predictions without Hospital Metrics')
axes[1].set_xlabel('Date')
#axes[1].legend()

# Ensure x-axis labels are horizontally aligned
for label in axes[1].get_xticklabels():
    label.set_rotation(0)

axes[1].legend()

# Common settings
plt.suptitle('Comparison of Prediction Models with and without Hospital Metrics')
#plt.xticks(rotation=90)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
# Save the plot as a PDF
plt.savefig("comparison_plot.pdf")
plt.show()
```
