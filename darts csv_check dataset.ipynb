```python
def load_dataset(name):
    if name == "EXOLGBM":
        df = pd.read_csv('./datasets/CAcases.csv')
        df.rename({'collection_date': 'timestamp', 'total_patients_hospitalized_covid_confirmed': 'y'}, axis='columns', inplace=True)
        exog_cols = ['timestamp','y','admissions_all_covid_confirmed', 'number_hospitals_reporting_today','admissions_70_covid_confirmed', 'inpatient_adult_beds']
        df = df[exog_cols]
        data = df.melt(id_vars=['timestamp'], value_vars=['admissions_all_covid_confirmed', 'number_hospitals_reporting_today','admissions_70_covid_confirmed','inpatient_adult_beds', 'y'], value_name='target')
        data.rename({'variable': 'item_id'}, axis='columns', inplace=True)
        data['timestamp'] = pd.to_datetime(data['timestamp'], format='%d-%m-%Y')
        data = data.pivot(columns="item_id", index="timestamp", values="target")
        data['y'] = data['y'].astype(float)
        data = data.interpolate()
        data.index = pd.to_datetime(data.index)
        return data
```
