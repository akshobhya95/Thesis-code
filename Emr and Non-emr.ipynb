```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import ShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder

# Assuming df is already loaded as per the given dataframe structure

# Define features and target variable
X = df.drop(columns=['jurisdiction', 'collection_date', 'state', 'total_patients_hospitalized_covid_confirmed'])
y = df['total_patients_hospitalized_covid_confirmed']

# Include random features
rng = np.random.RandomState(seed=42)
X["random_cat"] = rng.randint(3, size=X.shape[0])
X["random_num"] = rng.randn(X.shape[0])

# Define categorical and numerical columns
categorical_columns = X.select_dtypes(include=['object', 'category']).columns.tolist() + ['random_cat']
numerical_columns = X.select_dtypes(include=[np.number]).columns.tolist() + ['random_num']

# Define preprocessing pipeline
categorical_encoder = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)
numerical_pipe = SimpleImputer(strategy="mean")

preprocessing = ColumnTransformer(
    [
        ("cat", categorical_encoder, categorical_columns),
        ("num", numerical_pipe, numerical_columns),
    ],
    verbose_feature_names_out=True  # Set this to True to avoid name conflicts
)

# Define the pipeline with a random forest regressor
rf_regressor = Pipeline(
    [
        ("preprocess", preprocessing),
        ("regressor", RandomForestRegressor(n_estimators=100, random_state=42)),
    ]
)

# Perform ShuffleSplit cross-validation
shuffle_split = ShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
perm_importance = {}

for train_index, test_index in shuffle_split.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Train the random forest regressor
    rf_regressor.fit(X_train, y_train)
    
    # Calculate baseline mean squared error
    y_pred_baseline = rf_regressor.predict(X_test)
    baseline_mse = mean_squared_error(y_test, y_pred_baseline)
    
    # Calculate permutation feature importance
    for feature in X.columns:
        X_test_permuted = X_test.copy()
        X_test_permuted[feature] = np.random.permutation(X_test[feature])
        y_pred_permuted = rf_regressor.predict(X_test_permuted)
        perm_mse = mean_squared_error(y_test, y_pred_permuted)
        if feature in perm_importance:
            perm_importance[feature].append(baseline_mse - perm_mse)
        else:
            perm_importance[feature] = [baseline_mse - perm_mse]

# Average permutation importance over the splits
perm_importance_avg = {k: np.mean(v) for k, v in perm_importance.items()}
perm_importance_series = pd.Series(perm_importance_avg).sort_values(ascending=False)

# Display top 10 features for permutation importance
print("\nTop 20 features by permutation importance:")
print(perm_importance_series.head(20))

# Plot permutation feature importance
plt.figure(figsize=(10, 6))
perm_importance_series.head(20).sort_values().plot.barh()
plt.title("Top 20 Permutation Feature Importances")
plt.xlabel("Feature Importance")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import ShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder
from sklearn.inspection import permutation_importance

# Assuming df is already loaded and preprocessed

# Define features and target variable
X = df.drop(columns=['jurisdiction', 'collection_date', 'state', 'total_patients_hospitalized_covid_confirmed'])
y = df['total_patients_hospitalized_covid_confirmed']

# Include random features
rng = np.random.RandomState(seed=42)
X["random_cat"] = rng.randint(3, size=X.shape[0])
X["random_num"] = rng.randn(X.shape[0])

# Define categorical and numerical columns
categorical_columns = X.select_dtypes(include=['object', 'category']).columns.tolist() + ['random_cat']
numerical_columns = X.select_dtypes(include=[np.number]).columns.tolist() + ['random_num']

# Define preprocessing pipeline
categorical_encoder = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)
numerical_pipe = SimpleImputer(strategy="mean")

preprocessing = ColumnTransformer(
    [
        ("cat", categorical_encoder, categorical_columns),
        ("num", numerical_pipe, numerical_columns),
    ],
    verbose_feature_names_out=True  # Set this to True to avoid name conflicts
)

# Define the pipeline with a random forest regressor
rf_regressor = Pipeline(
    [
        ("preprocess", preprocessing),
        ("regressor", RandomForestRegressor(n_estimators=100, random_state=42)),
    ]
)

# Perform ShuffleSplit cross-validation
shuffle_split = ShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
perm_importance = {}

for train_index, test_index in shuffle_split.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Train the random forest regressor
    rf_regressor.fit(X_train, y_train)
    
    # Calculate baseline mean squared error
    y_pred_baseline = rf_regressor.predict(X_test)
    baseline_mse = mean_squared_error(y_test, y_pred_baseline)
    
    # Calculate permutation feature importance
    result = permutation_importance(rf_regressor, X_test, y_test, n_repeats=10, random_state=42, n_jobs=2)
    perm_importance_avg = result.importances_mean
    perm_importance_std = result.importances_std
    
    # Store the permutation importances
    perm_importance.update(dict(zip(X.columns, perm_importance_avg)))

# Feature importances from RandomForestRegressor
mdi_importances = pd.Series(rf_regressor.named_steps['regressor'].feature_importances_, 
                            index=rf_regressor.named_steps['preprocess'].get_feature_names_out())
mdi_importances_sorted = mdi_importances.sort_values(ascending=True)

# Plot impurity-based and permutation importances
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 8))

# Plot MDI (impurity-based) feature importance
mdi_importances_sorted.head(20).plot.barh(ax=ax1)
ax1.set_xlabel("Gini Importance")
ax1.set_title("Top 20 MDI Feature Importances")

# Plot permutation importance
perm_importance_series = pd.Series(perm_importance).sort_values(ascending=True)
perm_importance_series.head(20).plot.barh(ax=ax2)
ax2.set_xlabel("Decrease in Accuracy Score")
ax2.set_title("Top 20 Permutation Feature Importances")

fig.suptitle("Top 20 Features: Impurity-based vs. Permutation Importances")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.inspection import permutation_importance

# Define features and target variable
X = df.drop(columns=['jurisdiction', 'collection_date', 'state', 'total_patients_hospitalized_covid_confirmed' ,'proportion_inpatient_beds','proportion_prevalent_inpatient_beds'])
y = df['total_patients_hospitalized_covid_confirmed']

# Check and handle class distribution issues (if necessary)
class_counts = y.value_counts()
rare_classes = class_counts[class_counts < 2].index
X = X[~y.isin(rare_classes)]
y = y[~y.isin(rare_classes)]

# Define categorical and numerical columns
categorical_columns = [col for col in X.select_dtypes(include=['object', 'category']).columns]
numerical_columns = [col for col in X.select_dtypes(include=['number']).columns]

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Create preprocessing pipeline
categorical_encoder = OrdinalEncoder(
    handle_unknown="use_encoded_value", unknown_value=-1, encoded_missing_value=-1
)
numerical_pipe = SimpleImputer(strategy="mean")

preprocessing = ColumnTransformer(
    [
        ("cat", categorical_encoder, categorical_columns),
        ("num", numerical_pipe, numerical_columns),
    ],
    verbose_feature_names_out=False,
)

# Create and train the Random Forest model
rf = Pipeline(
    [
        ("preprocess", preprocessing),
        ("classifier", RandomForestClassifier(random_state=42)),
    ]
)
rf.fit(X_train, y_train)

# Compute MDI importances
feature_names = rf.named_steps['preprocess'].get_feature_names_out()
mdi_importances = pd.Series(
    rf.named_steps['classifier'].feature_importances_, index=feature_names
).sort_values(ascending=False)

# Extract and plot top 10 MDI importances
top_10_mdi = mdi_importances.head(10)

# Plot top 10 MDI importances using Matplotlib
plt.figure(figsize=(14, 10))
top_10_mdi.plot(kind='barh', color='skyblue')
plt.xlabel('Importance', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.title('Top 10 Random Forest Feature Importances (MDI)', fontsize=16)
plt.gca().invert_yaxis()
plt.gca().xaxis.set_tick_params(labelsize=12)
plt.gca().yaxis.set_tick_params(labelsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Compute permutation importances on test set
result = permutation_importance(
    rf, X_test, y_test, n_repeats=10, random_state=42, n_jobs=2
)

# Ensure correct handling of indices
importances_mean = result.importances_mean
importances_std = result.importances_std
sorted_idx = np.argsort(importances_mean)[::-1]  # Sort in decreasing order

# Extract top 10 Permutation importances
top_10_perm_importances = pd.DataFrame(
    {
        'Feature': feature_names[sorted_idx][:10],
        'Importance': importances_mean[sorted_idx][:10],
        'Importance_std': importances_std[sorted_idx][:10]
    }
)

# Plot top 10 Permutation importances using Matplotlib
plt.figure(figsize=(14, 10),dpi=200)
plt.barh(top_10_perm_importances['Feature'], top_10_perm_importances['Importance'], xerr=top_10_perm_importances['Importance_std'], color='lightcoral', capsize=5)
plt.xlabel('Importance', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.title('Top 10 Permutation Importances (test set)', fontsize=16 , fontweight='bold')
plt.gca().invert_yaxis()
plt.gca().xaxis.set_tick_params(labelsize=12)
plt.gca().yaxis.set_tick_params(labelsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the figure with increased DPI
plt.savefig('top_10_permutation_importances_test_set.png', dpi=300)  # Save with 200 DPI
plt.show()

# Optionally, also plot Permutation importances for the train set
result_train = permutation_importance(
    rf, X_train, y_train, n_repeats=10, random_state=42, n_jobs=2
)
importances_mean_train = result_train.importances_mean
importances_std_train = result_train.importances_std
sorted_idx_train = np.argsort(importances_mean_train)[::-1]  # Sort in decreasing order

top_10_perm_importances_train = pd.DataFrame(
    {
        'Feature': feature_names[sorted_idx_train][:10],
        'Importance': importances_mean_train[sorted_idx_train][:10],
        'Importance_std': importances_std_train[sorted_idx_train][:10]
    }
)

# Plot top 10 Permutation importances for train set using Matplotlib
plt.figure(figsize=(14, 10))
plt.barh(top_10_perm_importances_train['Feature'], top_10_perm_importances_train['Importance'], xerr=top_10_perm_importances_train['Importance_std'], color='lightgreen', capsize=5)
plt.xlabel('Importance', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.title('Top 10 Permutation Importances (train set)', fontsize=16)
plt.gca().invert_yaxis()
plt.gca().xaxis.set_tick_params(labelsize=50)
plt.gca().yaxis.set_tick_params(labelsize=50)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Sample data loading
date_range = pd.date_range(start='2020-08-01', end='2024-02-17', freq='D')
np.random.seed(0)
data = np.random.poisson(lam=50, size=len(date_range))
df = pd.DataFrame({'date': date_range, 'total_patients_hospitalized_covid_confirmed': data})
df.set_index('date', inplace=True)

# Calculate moving average (7-day moving average to smooth daily data)
df['hospitalized_avg'] = df['total_patients_hospitalized_covid_confirmed'].rolling(window=7).mean()

# Drop NaN values created by rolling window
df = df.dropna()

# Calculate z-scores
df['z_score_hospitalized'] = zscore(df['hospitalized_avg'])

# Define thresholds
emergency_threshold = 1.5
non_emergency_threshold = -0.5

# Classify periods into Emergency and Non-Emergency
df['classification'] = np.where(df['z_score_hospitalized'] > emergency_threshold, 'Emergency', 'Non-Emergency')

# Plot the results
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['hospitalized_avg'], label='Hospitalized (7-day MA)', color='blue')
plt.fill_between(df.index, 0, df['hospitalized_avg'],
                 where=df['classification'] == 'Emergency', color='red', alpha=0.3, label='Emergency')
plt.fill_between(df.index, 0, df['hospitalized_avg'],
                 where=df['classification'] == 'Non-Emergency', color='green', alpha=0.3, label='Non-Emergency')
plt.xlabel('Date')
plt.ylabel('Total Hospitalized (7-day MA)')
plt.title('Prevalent COVID-19 Hospitalizations: Emergency vs. Non-Emergency Periods in Califonia')
plt.legend()
plt.show()

# Display the classification
print(df[['hospitalized_avg', 'z_score_hospitalized', 'classification']])
```

```python
# Example of normalizing a specific column before calculation
from sklearn.preprocessing import StandardScaler

# Assuming 'total_patients_hospitalized_covid_confirmed' needs normalization
scaler = StandardScaler()
df['normalized_hospitalized'] = scaler.fit_transform(df[['admissions_all_covid_confirmed']])

# Convert collection_date to datetime format
df['collection_date'] = pd.to_datetime(df['collection_date'], format='%d-%m-%Y')

# Make a copy of the original dataframe
df_copy = df.copy()

# Calculate moving average (7-day moving average to smooth daily data)
df_copy['hospitalized_avg'] = df_copy['normalized_hospitalized'].rolling(window=7).mean()

# Drop NaN values created by rolling window
df_copy.dropna(inplace=True)

# Calculate median and median absolute deviation (MAD)
median = np.median(df_copy['hospitalized_avg'])
mad = median_abs_deviation(df_copy['hospitalized_avg'])

# Calculate z-scores using median and MAD
df_copy['z_score_hospitalized'] = (df_copy['hospitalized_avg'] - median) / mad

# Define thresholds
emergency_threshold = 3
non_emergency_threshold = -3

# Classify periods into Emergency and Non-Emergency
df_copy['classification'] = np.where(df_copy['z_score_hospitalized'] > emergency_threshold, 'Emergency', 'Non-Emergency')

# Plot the results using df_copy (plotting uses normalized data)
plt.figure(figsize=(14, 7))
plt.plot(df_copy['collection_date'], df_copy['hospitalized_avg'], label='Hospitalized (7-day MA)', color='blue')
plt.fill_between(df_copy['collection_date'], 0, df_copy['hospitalized_avg'],
                 where=df_copy['classification'] == 'Emergency', color='red', alpha=0.3, label='Emergency')
plt.fill_between(df_copy['collection_date'], 0, df_copy['hospitalized_avg'],
                 where=df_copy['classification'] == 'Non-Emergency', color='green', alpha=0.3, label='Non-Emergency')
plt.xlabel('Date')
plt.ylabel('Total Hospitalized (7-day MA)')
plt.title(f'Newer COVID-19 Hospitalizations: Emergency vs. Non-Emergency Periods in Califonia (CA) with \nemergency threshold={emergency_threshold:.2f} ,\nMedian={median:.2f}, MAD={mad:.2f}', fontsize=16)  # Include median and MAD in the title
plt.legend()
plt.show()

# Merge columns from df_copy back into df based on date index
df = pd.merge(df, df_copy[['hospitalized_avg', 'z_score_hospitalized', 'classification']],
              how='left', left_index=True, right_index=True, suffixes=('_x', ''))

# Display the classification in the original dataframe
print(df[['hospitalized_avg', 'z_score_hospitalized', 'classification']])
```
